2. Generation Counts

```python
def gen_count(s):
    generation_counts={}
    for p in s:
        if p['generation'] in generation_counts:
            generation_counts[p['generation']] += 1
        else:
            generation_counts[p['generation']] =1
    return generation_counts
gen_count(data)
```

3. Mean Weights by Primary Type

```python
def gen_count(s):
    newdic={}
    for p in s:
        if p['primary_type'] in newdic:
            if p['weight_kg']!= None:
                newdic[p['primary_type']] += p['weight_kg']
        else:
            if p['weight_kg']!= None:
                newdic[p['primary_type']] = p['weight_kg']
    return newdic
def gen_counts(n):       
    newdic1={}
    for p in n:
        if p['primary_type'] in newdic1:
            if p['weight_kg']!= None:
                newdic1[p['primary_type']] += 1
        else:
            if p['weight_kg']!= None:
                newdic1[p['primary_type']] = 1
    return newdic1

def whole_fun():
    a=gen_count(data)
    b=gen_counts(data)
    primary_weights= dict((k,float(a[k])/b[k])for k in a)
    return primary_weights
whole_fun()
```

4. Filter by Hit Points

```python
def filter_hp(s,min_hp,max_hp):
    L= []
    for i in s:
        if i['hp'] in range(min_hp,max_hp+1):
            L.append(((i['hp']),(i['name'])))
    L.sort()
    return L
filter_hp(data,1,255)
    
```

5. Median Attack & Defense by Generation

```python
def median(info):
    info.sort()
    if len(info) % 2 != 0:
        avg= info[len(info) // 2]
    else:
        avg=(info[(len(info) // 2)-1] + info[len(info) // 2]) / 2
    return avg
def final_dictionaries(sn):
    nd={}
    for i in sn:
        if not i['generation'] in nd.keys():
            
            nd[i['generation']]['sa'].append(i['attack'])
            nd[i['generation']]['sd'].append(i['defense'])
    gen_atk_def= []
    for i, b in nd.items():
        r=median(b['sa'])
        c=median(b['sd'])
        gen_atk_def.append({'generation':i, 'sa':r, 'sd':c})
    return gen_atk_def
final_dictionaries(data)
```
