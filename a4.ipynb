```python
# Do not modify this cell, just run it to download the file

from urllib.request import urlretrieve

url = "http://faculty.cs.niu.edu/~dakoop/cs503-2021sp/a4/illinois-climate-monthly.txt"
local_fname = "illinois-climate-monthly.txt"
urlretrieve(url, local_fname)
```

1. Read Data

```python
data = []
a = open("illinois-climate-monthly.txt")
b=iter(a)
key=next(b).split()
for l in b:
    data.append(dict(zip(key,l.split())))
data    
```

2. Split Year and Month

```python
for date in data:
    date['year']=date['YearMonth'][-6:-2]
    date['month']=date['YearMonth'][-2:]
data
```

3. Compute Yearly Values

```python
info = sorted(set(value['year'] for value in data))
year_data =[]
for year in info:
    Spcp = [float(value['PCP']) for value in data if value['year']== year]
    Atavg = [float(value['TAVG']) for value in data if value['year'] == year]
    Apdsi = [float(value['PDSI']) for value in data if value['year'] == year]
    a=sum((Spcp))
    b=sum(Atavg)/len(Atavg)
    c=sum(Apdsi)/len(Apdsi)
    year_data.append({'year': int(year),'PCP': a ,'TAVG':b,'PDSI' : c })
year_data
```
